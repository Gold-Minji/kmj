#인공지능사관학교 1주차 과제

인공지능의 발전을 알 수 있는 것은 번역이라고 본 강의에서 들었다.

번역기로 널리 쓰이고 있는 구글 번역기가 있고, 네이버 번역기인 파파고가 있다는 것은 익히 알고 있었다.

하지만, 비장애인이 쓰는 언어만 번역이 되고 장애인이 사용하고 있는 언어인 수어는 통역해 주지 못한다는 점이 못내 아쉬웠다.

필요성에 대해서는 BBC NEWS KOREA의 [코로나19가 사회적 약자에게 미친 영향 4가지](https://www.bbc.com/korean/features-51797503) 를 확인해 보면 된다.

이 기사를 보면 문자로 의사소통을 하면 되지 왜 굳이 수어로 통역을 해야하는지에 대한 답을 얻을 수 있다.

기사를 보면, 

' _"자막방송이라도 의지하면 되지 않나 생각하는 분들도 적지 않으신데 한국수어와 한국어는 비슷해 보이지만 엄연히 다른 언어입니다. 
문법 규칙부터가 다르고 같은 단어라 해도 서로 다른 뉘앙스가 담겨 있어 일상에서 오해의 소지가 종종 있어요."_

한국수어를 모국어로 하는 청각장애인들에게는 자막방송이나 한국어로 된 기사보다 수어 통역이 절실할 수밖에 없다는 얘기다.'

라는 인터뷰가 있으며, 이로인해 **청각장애인들이 정보를 전달받는 데에 얼마나 소외되어있었는지** 알 수 있다.

또한 코로나로 인해서 사이버강의로 전환되었지만, ["입모양만 보고 온라인 강의 들어요", 온라인 강의에서 소외되는 청각장애인 학생들](https://www.youtube.com/watch?v=iHgDgymoGNs)이라는 SBS 뉴스에서도 알 수 있는 것처럼 많은 곳에서 비장애인들이 의도하지 않았음에도 차별받았음을 알 수 있었다.


그래서 한국어와 한국수어에 대한 통역의 필요성을 느꼈으며, 수어통역에 관련된 자료를 찾아보았다. 

1. 한국어 음성을 인식해 문자로 바꾸기 위해서 사용되는 음성인식 알고리즘에 대해 알아보았다.
    1. HMM( Hidden Markov Model): 은닉 마코프 모델은 기존 마코프 모델에서 상태정보가 은닉된 통계적 모델이다. 이 모델은 대량으로 출력된 데이터를 통계적으로 패턴 분석하여 입력된 정보를 추론하는데에 주로 응용된다.
    2. Fully connected DNN(Deep Neyral Network)
    3. LSTM(Long Short-Term Memory) : 시간순서로 
  음성인식에 사용되는 학습모델은 크게 음향모델(Acoustic Model)과 언어모델(Language Model)로 나누어져 있다. 음향모델은 음향적 특성을 통계적으로 모델링하여 학습을 하는 것으로, 음석인식 엔진에서 제공하는 기본 모델을 기반으로 실제 적용하고자 하는 음성의 특성을 추가하는 적응학습이 가능하다.

  이런 알고리즘 들을 통해서 음성인식을 할 때 단어를 대응시켜서 번역을 하는 것이 아니라 단어간의 관계, 문맥 등을 파악해서 번역 하고, 이런 작업을 계속하며 학습을 해서 더 나은 결과를 얻는다.


2. 다음으로는 텍스트를 인식하는 것으로 알아보았다. 한글 텍스트를 인식하는 것으로는 네이버 파파고 NMT에 대해 간단한 설명을 보았다.
Neural Machine Translation은 인공신경망 기계번역이다. 번역하기 위해 입력된 문장은 인공신경망이라는 가상공간에 벡터값으로 바뀌어 저장이 되고,
비슷한 맥락을 가진 문장들이 가까이 위치해 입력된 문장 주변에 있는 정보들을 참고하여, 더 자연스러운 번역을 하고, Papago NMT는 네이버 서비스를 통해
쌓인 대규모 데이터를 매일 자동으로 학습하며 이런 번역모델을 통해 기존의 SMT방식의 번역보다 정확한 결과를 제공한다.

3. 이미지 인식의 한 예로써는 입술모양(이미지)을 읽어서 말을 이해하는, 구순술에 적용된 예를 살펴보았다. [Zhao, Ya & Xu, Rui & Wang, Xinchao & Hou, Peng & Tang, Haihong & Song, Mingli. (2019). Hearing Lips: Improving Lip Reading by Distilling Speech Recognizers.](https://arxiv.org/pdf/1911.11502.pdf) 이 논문에서 보아야 할 부분은 3번째 장에서 Proposed Method 에서 speech recognizer 는 attention기반의 seq2seq 구조를 사용한 것을 알 수 있다.
또한 범죄에 관련되어서 유명해진 딥 페이크는 이미지를 '감별'하는 알고리즘과 '생성'하는 알고리즘으로 구성된 인공지능(AI) 기술로 만든다. 두 알고리즘을 대조해 원본과의 오차를 줄이는 방법으로 가짜를 만들어내는 것이다. 수어에서는 표정또한 말을 전달하는데에 필요한 부분이기 때문에, 이를 이용해보려 한다. 


4. 자율주행에서는 라이다 센서를 이용해서 3D로 주변 상황을 그린 다음에 사용하는것을 구글, 현대 등 자율주행을 도입하려는 많은 회사에서 사용을 하고 있다. 하지만 테슬라는 라이다를 사용하지 않고 카메라로 이미지로 장애물들을 인식을해서 자율주행을 수행한다. 자율주행에 대한 많은 연구와 발전이 진행되면서 많은 오픈소스들이 제공되어 있기도 하다. 자율주행 알고리즘을 검증하고 고도화시키는건 연구하는 사람들의 몫이라고 생각한다.

수어통역에 관해 조사를 하였으며, 음성인식, 텍스트 인식, 이미지처리, 자율주행, 이 4가지의 AI가 쓰이는 대표적인 항목에 대해서도 간단한 조사를 해 보았다. 이 중 청각장애인들을 위한 번역기를 만드는 데에 필요한 부분이 있기도 했지만 라이다센서를 이용한 자율주행 처럼 필요 없는 부분이 있기도 하였다.

**하지만  무엇보다 지금의 나의 지식과 능력으로써는 아직 목표를 이루기 위해서는 배워야 할 점이 너무 많다는 것을 이번 조사 과제를 통해 알게 되었다.**
